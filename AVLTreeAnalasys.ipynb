# AVL bench marks

```python
import numpy as np
import pandas as pd

AVL_df1 = pd.read_csv('csv/AVL_average_insert_benchmark.csv')
AVL_df2 = pd.read_csv('csv/AVL_average_search_benchmark.csv')
AVL_df3 = pd.read_csv('csv/AVL_average_delete_benchmark.csv')
AVL_df4 = pd.read_csv('csv/AVL_WorstCase_0n_insert_benchmark.csv')
AVL_df5 = pd.read_csv('csv/AVL_WorstCase_0n_search_benchmark.csv')
AVL_df6 = pd.read_csv('csv/AVL_WorstCase_0n_delete_benchmark.csv')
AVL_df7 = pd.read_csv('csv/AVL_WorstCase_n0_insert_benchmark.csv')
AVL_df8 = pd.read_csv('csv/AVL_WorstCase_n0_search_benchmark.csv')
AVL_df9 = pd.read_csv('csv/AVL_WorstCase_n0_delete_benchmark.csv')

sequence_interval = 0.01
downsampled_interval = 10
step_size = np.round(downsampled_interval / sequence_interval).astype("int")

# downsampled_df
AVL_df1 = AVL_df1.iloc[::step_size, :]
AVL_df2 = AVL_df2.iloc[::step_size, :]
AVL_df3 = AVL_df3.iloc[::step_size, :]
AVL_df4 = AVL_df4.iloc[::step_size, :]
AVL_df5 = AVL_df5.iloc[::step_size, :]
AVL_df6 = AVL_df6.iloc[::step_size, :]
AVL_df7 = AVL_df7.iloc[::step_size, :]
AVL_df8 = AVL_df8.iloc[::step_size, :]
AVL_df9 = AVL_df9.iloc[::step_size, :]

#df.info()
import matplotlib.pyplot as plt

plt.plot(AVL_df1["Number of Inserts"], AVL_df1["Elapsed Time (ms)"], color='Red', linestyle='solid', label='insert (AVG)')
plt.plot(AVL_df2["Number of Inserts"], AVL_df2["Elapsed Time (ms)"], color='Green', linestyle='solid', label='search (AVG)')
plt.plot(AVL_df3["Number of Inserts"], AVL_df3["Elapsed Time (ms)"], color='Blue', linestyle='solid', label='delete (AVG)')
plt.plot(AVL_df4["Number of Inserts"], AVL_df4["Elapsed Time (ms)"], color='Red', linestyle='dashed', label='insert (WC)')
plt.plot(AVL_df5["Number of Inserts"], AVL_df5["Elapsed Time (ms)"], color='Green', linestyle='dashed', label='search (WC)')
plt.plot(AVL_df6["Number of Inserts"], AVL_df6["Elapsed Time (ms)"], color='Blue', linestyle='dashed', label='delete (WC)')
plt.plot(AVL_df7["Number of Inserts"], AVL_df7["Elapsed Time (ms)"], color='Red', linestyle='dotted', label='insert (WC)')
plt.plot(AVL_df8["Number of Inserts"], AVL_df8["Elapsed Time (ms)"], color='Green', linestyle='dotted', label='search (WC)')
plt.plot(AVL_df9["Number of Inserts"], AVL_df9["Elapsed Time (ms)"], color='Blue', linestyle='dotted', label='delete (WC)')
plt.title("AVL tree benchmarkings")
plt.xlabel("Number of operations")
plt.ylabel("Elapsed Time (ms)")
plt.legend()

plt.tight_layout()
plt.savefig("doc/images/AVLTreeBenchmarkings.png")
plt.show()
```

# RedBlack tree bench marks

```python
import numpy as np
import pandas as pd

RB_df1 = pd.read_csv('csv/RedBlack_average_insert_benchmark.csv')
RB_df2 = pd.read_csv('csv/RedBlack_average_search_benchmark.csv')
RB_df3 = pd.read_csv('csv/RedBlack_average_delete_benchmark.csv')
RB_df4 = pd.read_csv('csv/RedBlack_WorstCase_0n_insert_benchmark.csv')
RB_df5 = pd.read_csv('csv/RedBlack_WorstCase_0n_search_benchmark.csv')
RB_df6 = pd.read_csv('csv/RedBlack_WorstCase_0n_delete_benchmark.csv')
RB_df7 = pd.read_csv('csv/RedBlack_WorstCase_n0_insert_benchmark.csv')
RB_df8 = pd.read_csv('csv/RedBlack_WorstCase_n0_search_benchmark.csv')
RB_df9 = pd.read_csv('csv/RedBlack_WorstCase_n0_delete_benchmark.csv')

sequence_interval = 0.01
downsampled_interval = 10
step_size = np.round(downsampled_interval / sequence_interval).astype("int")

# downsampled_RB_df
RB_df1 = RB_df1.iloc[::step_size, :]
RB_df2 = RB_df2.iloc[::step_size, :]
RB_df3 = RB_df3.iloc[::step_size, :]
RB_df4 = RB_df4.iloc[::step_size, :]
RB_df5 = RB_df5.iloc[::step_size, :]
RB_df6 = RB_df6.iloc[::step_size, :]
RB_df7 = RB_df7.iloc[::step_size, :]
RB_df8 = RB_df8.iloc[::step_size, :]
RB_df9 = RB_df9.iloc[::step_size, :]

#RB_df.info()
import matplotlib.pyplot as plt

plt.plot(RB_df1["Number of Inserts"], RB_df1["Elapsed Time (ms)"], color='darkviolet', linestyle='solid', label='insert')
plt.plot(RB_df2["Number of Inserts"], RB_df2["Elapsed Time (ms)"], color='darkkhaki', linestyle='solid', label='search')
plt.plot(RB_df3["Number of Inserts"], RB_df3["Elapsed Time (ms)"], color='darkorange', linestyle='solid', label='delete')
plt.plot(RB_df4["Number of Inserts"], RB_df4["Elapsed Time (ms)"], color='darkviolet', linestyle='dashed', label='insert')
plt.plot(RB_df5["Number of Inserts"], RB_df5["Elapsed Time (ms)"], color='darkkhaki', linestyle='dashed', label='search')
plt.plot(RB_df6["Number of Inserts"], RB_df6["Elapsed Time (ms)"], color='darkorange', linestyle='dashed', label='delete')
plt.plot(RB_df7["Number of Inserts"], RB_df7["Elapsed Time (ms)"], color='darkviolet', linestyle='dotted', label='insert')
plt.plot(RB_df8["Number of Inserts"], RB_df8["Elapsed Time (ms)"], color='darkkhaki', linestyle='dotted', label='search')
plt.plot(RB_df9["Number of Inserts"], RB_df9["Elapsed Time (ms)"], color='darkorange', linestyle='dotted', label='delete')
#plt.yscale("log")
plt.title("Red-Black tree benchmarkings")
plt.xlabel("Number of operations")
plt.ylabel("Elapsed Time (ms)")
plt.legend()

plt.tight_layout()
plt.savefig("doc/images/RedBlackTreeBenchmarkings.png")
plt.show()
```

# HashTable Benchmark

```python
import numpy as np
import pandas as pd

Hash_df1 = pd.read_csv('csv/HashTable_insert_benchmark.csv')
Hash_df2 = pd.read_csv('csv/HashTable_get_benchmark.csv')
Hash_df3 = pd.read_csv('csv/HashTable_remove_benchmark.csv')


sequence_interval = 0.01
downsampled_interval = 10
step_size = np.round(downsampled_interval / sequence_interval).astype("int")

# downsampled_df
Hash_df1 = Hash_df1.iloc[::step_size, :]
Hash_df2 = Hash_df2.iloc[::step_size, :]
Hash_df3 = Hash_df3.iloc[::step_size, :]


#df.info()
import matplotlib.pyplot as plt

plt.plot(Hash_df1["Number of Inserts"], Hash_df1["Elapsed Time (ms)"], color='black', linestyle='solid', label='insert (AVG)')
plt.plot(Hash_df2["Number of Inserts"], Hash_df2["Elapsed Time (ms)"], color='cadetblue', linestyle='solid', label='get (AVG)')
plt.plot(Hash_df3["Number of Inserts"], Hash_df3["Elapsed Time (ms)"], color='Slategrey', linestyle='solid', label='remove (AVG)')

#plt.yscale("log")
plt.title("HashTable benchmarkings")
plt.xlabel("Number of operations")
plt.ylabel("Elapsed Time (ms)")
plt.legend()

plt.tight_layout()
plt.savefig("doc/images/HashTableBenchmarkings.png")
plt.show()
```
