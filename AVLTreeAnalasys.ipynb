# Insert Benchmark

```python

import numpy as np
import pandas as pd

df = pd.read_csv('insert_benchmark.csv')

sequence_interval = 0.01
downsampled_interval = 10
step_size = np.round(downsampled_interval / sequence_interval).astype("int")

# downsampled_df
df = df.iloc[::step_size, :]
#df.info()
```

```python
import matplotlib.pyplot as plt

plt.plot(df["Number of Inserts"], df["Elapsed Time (ms)"])
plt.xlabel("Number of Inserts")
plt.ylabel("Elapsed Time (ms)")
plt.show()
```
